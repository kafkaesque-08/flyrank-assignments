## 1. My lane as an ML task (type)

Task Type: Unsupervised Machine Learning – Clustering.

Description: This is a partitioning problem where we do not have historical, human-labeled target classes for what makes a page a specific "archetype." Instead, we use distance-based vectors to group multi-dimensional content footprints into distinct operational profiles based on similarities in performance and structural metadata.

## 2. Target or proxy

Target Definition: Because this is an unsupervised clustering task, there is no single target column to predict. Instead, the "target" is the Cluster Assignment ID derived from the resulting cluster centroids.

The algorithm will calculate assignments based on feature boundaries across three main signal groups: traffic volume variance (impressions_90d, clicks_90d), engagement behavior (engagement_rate, scroll_rate), and content attributes (content_age_days, word_count).

## 3. Success metric

Mathematical Metric (Intrinsic): Silhouette Score and Within-Cluster Sum of Squares (WCSS / Inertia). We will use these to evaluate cluster compactness and separation density, ensuring the mathematical boundaries are distinct.

## 5. Why ML beats a fixed rule here

A fixed heuristic rule (like if impressions > 5000 then high_value) fails because it only looks at a single dimension and handles massive outliers poorly. In our dataset, traffic spans from 0 to over 500,000 impressions with extreme variance.
